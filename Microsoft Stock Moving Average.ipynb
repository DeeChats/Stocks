```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
#Importing Dataset
data=pd.read_csv("C:/Users/deeks/Downloads/Microsoft_Stock.csv")
data
```

```python
#3 yearly Moving Average
data['SMA_3']= data['Close'].rolling(window=10).mean()
print(data['SMA_3'].tail(30))
```

```python
#line plot
data['Close'].plot(color='blue',linewidth=3,figsize=(12,6))
data['SMA_3'].plot(color='yellow',linewidth=3,figsize=(12,6))
#title and labels
plt.title('The Microsoft Closing Stock Price',fontsize=20)
plt.xlabel('Date',fontsize=16)
plt.ylabel('Closing Stock Price',fontsize=16)
```
